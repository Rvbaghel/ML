```python
import numpy as np
```

```python
class ActivationModuleClass:
    def __init__(self, x_data):
        self.x = x_data


    def call_sigmoid(self):
        return 1 / (1 + np.exp(-self.x))
    
    def call_sigmoid_derivative(self):
        s = self.call_sigmoid()
        return s * (1 - s)
    

    def call_relu(self):
        return np.maximum(0, self.x)
    
    def call_relu_derivative(self):
        return np.where(self.x > 0, 1, 0)
    

    def call_leaky_relu(self, a=0.01):
        return np.where(self.x > 0, self.x, a * self.x)
    
    def call_leaky_relu_derivative(self, a=0.01):
        return np.where(self.x > 0, 1, a)
    

    def call_tanh(self):
        return np.tanh(self.x)
    
    def call_tanh_derivative(self):
        return 1 - np.tanh(self.x) ** 2
    


    def call_softmax(self):
        exps = np.exp(self.x - np.max(self.x))  
        return exps / np.sum(exps)
    
    def call_softmax_derivative(self):
        
        s = self.call_softmax().reshape(-1, 1)
        return np.diagflat(s) - np.dot(s, s.T)


```

```python
x_test = np.array([-2, -1, 0, 1, 2])
act = ActivationModuleClass(x_test)
print(f"Data points {x_test}")
print("Sigmoid:", act.call_sigmoid())
print("Sigmoid Derivative:", act.call_sigmoid_derivative())
print("ReLU:", act.call_relu())
print("ReLU Derivative:", act.call_relu_derivative())
print("Leaky ReLU:", act.call_leaky_relu())
print("Leaky ReLU Derivative:", act.call_leaky_relu_derivative())
print("Tanh:", act.call_tanh())
print("Tanh Derivative:", act.call_tanh_derivative())


x_vec = np.array([1.0, 2.0, 3.0])
soft_act = ActivationModuleClass(x_vec)
print("Softmax:", soft_act.call_softmax())
print("Softmax Derivative (Jacobian):\n", soft_act.call_softmax_derivative())
```
