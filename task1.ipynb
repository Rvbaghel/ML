```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

# Parameters
n = 100  # Total tosses
k = 60   # Number of heads
p_values = np.linspace(0, 1, 100)  # 100 values between 0 and 1

# Calculate Likelihood for each p
# L(p) = (n choose k) * p^k * (1-p)^(n-k)
likelihood = comb(n, k) * (p_values**k) * ((1 - p_values)**(n - k))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(p_values, likelihood, color='blue', lw=2)
plt.axvline(x=0.6, color='red', linestyle='--', label='MLE (p=0.6)')
plt.title('Likelihood Function for Binomial Distribution (60 Heads in 100 Tosses)')
plt.xlabel('Probability of Head (p)')
plt.ylabel('Likelihood')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
