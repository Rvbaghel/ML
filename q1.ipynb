```python
import numpy as np
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,

)
```

```python
def apply_sigmoid(z):
    return 1 / (1 + np.exp(-z))
    
```

```python
def calc_loss(z):

    loss = -  np.sum( (y * np.log(z)) + ((1-y) * np.log(1-z)) )
    return   loss / X.shape[0]


def logistic_regression():
    
    epochs = 5000
    w = np.random.randn(2)
    b = np.random.randn(1)
    m,n = X.shape
    lr = 0.01
    loss = None
    
    for epoch in range(epochs):
        
        z = apply_sigmoid(np.dot(w, X.T) + b).T
        
            
        loss = calc_loss(z)
        if epoch == 0:
            
            print("initial loss", loss)
        err = z - y
        dw = (1/m) * (np.dot(X.T, err))
        w = w - (lr * dw).T
        
        db = (1/m) * np.sum(err)
        b = b - (lr  * db)
        
        
    print(f"final loss after {epochs} epochs: {loss}")
        
        
        
        
    
    


logistic_regression()
```
