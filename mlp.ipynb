```python
import numpy as np

X = np.array([
    [0.1, 0.8],
    [0.9, 0.1],
    [0.2, 0.9],
    [0.8, 0.2],
    [0.4, 0.5],
    [0.6, 0.4],
    [0.1, 0.2],
    [0.9, 0.9]
])
y = np.array([[1], [0], [1], [0], [1], [0], [0], [1]])

print(X,y)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))


def derivative(x):
    s=sigmoid(x)
    return s*(1-s)

w1=np.random.randn(4,2)
b1=np.zeros((4,1))

w2=np.random.randn(4,4)
b2=np.zeros((4,1))

w3 = np.random.randn(1, 4)
b3 = np.zeros((1, 1))

lr=0.1
target=y
       
for i in range(2000):
        hl_1=np.dot(w1,X.T)+b1
        a1=sigmoid(hl_1)

        hl_2=np.dot(w2,a1)+b2
        a2=sigmoid(hl_2)
        
        hl_3=np.dot(w3,a2)+b3
        a3=sigmoid(hl_3)
        
        #backword
        loss=-np.sum(target.T*np.log(a3)+(1-target.T)*np.log(1-a3))
        #layer 3
        dl_a3=a3-target.T
        dw3=np.dot(dl_a3,a2.T)
        db3 = np.sum(dl_a3, axis=1, keepdims=True)

        #layer 2 
        dl_a2 = np.dot(w3.T, dl_a3) * derivative(hl_2) 
        dw2 = np.dot(dl_a2, a1.T)
        db2 = np.sum(dl_a2, axis=1, keepdims=True)

        #layer 3
        dl_a1 = np.dot(w2.T, dl_a2) * derivative(hl_1)
        dw1 = np.dot(dl_a1, X)
        db1 = np.sum(dl_a1, axis=1, keepdims=True)

        w3-=lr*dw3
        b3-=lr * db3
        w2 -=lr * dw2
        b2 -=lr * db2
        w1 -=lr * dw1
        b1 -=lr * db1
        if i % 100 == 0:
             print(f"Iteration {i}, Loss: {loss}")
```
